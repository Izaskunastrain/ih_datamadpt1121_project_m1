```python
df_bicimad_original = pd.read_csv('../data/bicimad_stations.csv')
df_bicimad = df_bicimad_original [['name', 'address', 'geometry_coordinates']]

def lat (x):
    b= x.split(',')
    c= b[0].replace('[','')
    d= float(c)
    return d
    
def long (x):
    b= x.split(',')
    c= b[1].replace(']','')
    d= float(c)
    return d
```

```python
df_bicimad['Longitude.bicimad'] = df_bicimad.apply(lambda x: lat(x['geometry_coordinates']),axis=1)
df_bicimad['Latitude.bicimad'] = df_bicimad.apply(lambda x: long(x['geometry_coordinates']),axis=1)
df_bicimad
```

```python
df_bicimad_final= df_bicimad[['name', 'address', 'Latitude.bicimad', 'Longitude.bicimad']]
```

```python
# Distance function
```

```python
def to_mercator(lat, long):
    # transform latitude/longitude data in degrees to pseudo-mercator coordinates in metres
    c = gpd.GeoSeries([Point(lat, long)], crs=4326)
    c = c.to_crs(3857)
    return c
```

```python
def distance_meters(location_centros, location_bicimad):
    # return the distance in metres between to latitude/longitude pair point in degrees (i.e.: 40.392436 / -3.6994487)
    return location_centros.distance(location_bicimad)
```
