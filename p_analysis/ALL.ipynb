```python
df_bicimad_original = pd.read_csv('../data/bicimad_stations.csv')
df_bicimad = df_bicimad_original [['name', 'address', 'geometry_coordinates']]

def lat (x):
    b= x.split(',')
    c= b[0].replace('[','')
    d= float(c)
    return d
    
def long (x):
    b= x.split(',')
    c= b[1].replace(']','')
    d= float(c)
    return d
```

```python
df_bicimad['Latitude'] = df_bicimad.apply(lambda x: lat(x['geometry_coordinates']),axis=1)
df_bicimad['Longitude'] = df_bicimad.apply(lambda x: long(x['geometry_coordinates']),axis=1)
df_bicimad
```
