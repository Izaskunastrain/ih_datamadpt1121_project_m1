#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(6,3))
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT',fontweight='bold',fontsize=20 ,  horizontalalignment='center',)
ax1.set_xlabel('x label' ,fontweight='heavy' )
ax1.set_ylabel('y label',fontweight='heavy')

ax2.plot(x,z)
ax2.set_title('QUADRATIC PLOT',fontweight='bold' ,fontsize=20,  horizontalalignment='center')
ax2.set_xlabel('x label',fontweight='ultralight')
ax2.set_ylabel('y label',fontweight='ultralight')
fig.tight_layout()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100)
e= math.e

y= x*x
z= e**x
```

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(6,3))
ax1.plot(x,y)
ax1.set_title('SQUARE X',fontweight='bold',fontsize=20 ,  horizontalalignment='center',)
ax1.set_xlabel('x label' ,fontweight='ultralight' )
ax1.set_ylabel('y label',fontweight='ultralight')
ax1.set_xscale('linear')
ax1.set_yscale('linear')

ax2.plot(x,z)
ax2.set_title('E ELEVATED X',fontweight='bold' ,fontsize=20,  horizontalalignment='center')
ax2.set_xlabel('x label',fontweight='ultralight')
ax2.set_ylabel('y label',fontweight='ultralight')
ax2.set_xscale('log')
ax2.set_yscale('log')
fig.tight_layout()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(6,3))
ax1.plot(x,y, label="y= x*x")
ax1.set_title('SQUARE X',fontweight='bold',fontsize=20 ,  horizontalalignment='center',)
ax1.set_xlabel('x label' ,fontweight='ultralight' )
ax1.set_ylabel('y label',fontweight='ultralight')
ax1.set_xscale('linear')
ax1.set_yscale('linear')
ax1.legend(loc='upper left')

ax2.plot(x,z, label="y= e**x")
ax2.set_title('E ELEVATED X',fontweight='bold' ,fontsize=20,  horizontalalignment='center')
ax2.set_xlabel('x label',fontweight='ultralight')
ax2.set_ylabel('y label',fontweight='ultralight')
ax2.set_xscale('log')
ax2.set_yscale('log')
fig.tight_layout()
ax2.legend(loc='upper left')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, axes = plt.subplots(1, 2, figsize=(12,4))
fitbit.sort_values(by=["Steps", "Months_encoded", "Work_or_Weekend"], inplace=True)

sns.scatterplot(ax=axes[0],x="Months_encoded", y="Steps", data=fitbit)
axes[0].set_title('MEDIAN STEPS BY MONTH_ENCODED')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
sns.barplot(ax=axes[1],x="Months_encoded", y="Steps",hue='Work_or_Weekend', data=fitbit)
axes[1].set_title('MEDIAN STEPS BY WORK_OR_WEEKEND')
labels= ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
axes[1].set_xticklabels(labels, rotation = 45)

fig.savefig("./../figures/fitbit.png") 

```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

plots= ['Minutes Lightly Active', 'Minutes Very Active','Minutes Sedentary' ]

fig, axes = plt.subplots(1, 3, figsize=(12,4))

for i in range(3):
    sns.scatterplot(ax=axes[i],x=plots[i], y="Steps", data=fitbit)
    axes[i].set_title(plots[i])

```
